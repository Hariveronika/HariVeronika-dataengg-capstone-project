```python
def is_valid_email(email):
    if pd.isna(email): return False
    pattern =r"^[\w\.-]+@[\w\.-]+\.\w+$"
    return bool(re.match(pattern, str(email)))

emp_df['EmailValid'] = emp_df['Email'].apply(is_valid_email)
```

```python
# #In data analysis, we often need to identify outliers — values that are far from the rest.
# #Two popular methods to detect outliers are Z-score and IQR 


# # Z-Score Method
# Z-score tells how far a data point is from the mean in terms of standard deviation.

# Formula: 
# 𝑍= data point(X)-mean/ standard deviation
# A Z-score > 3 or < -3 is usually considered an outlier.
# it works best when the data is normally distributed.


# IQR looks at the spread of the middle 50% of the data.

# Formula: IQR = Q3 - Q1
# Outliers are values that fall below 
# This method is useful for skewed data and is not affected by extreme values.#
```

```python
print(emp_df)
```

```python
# salary_data=emp_df['Age'].dropna()
# Q1 = salary_data.quantile(0.25)
# Q3 = salary_data.quantile(0.75)

# IQR = Q3 - Q1

# lower_bound = Q1 - 1.5 * IQR
# upper_bound = Q3 + 1.5 * IQR

# outliers = emp_df[(emp_df['Age'] < lower_bound) | (emp_df['Age'] > upper_bound)]

# print("Outliers in age column:")
# print(outliers[['Age']]) 
```

```python
# Calculate z-scores manually
emp_df['z_Age'] = (emp_df['Age'] - emp_df['Age'].mean()) / emp_df['Age'].std()
emp_df['z_Salary'] = (emp_df['Salary'] - emp_df['Salary'].mean()) / emp_df['Salary'].std()

# Find outliers
age_outliers = emp_df[emp_df['z_Age'].abs() > 3]
sal_outliers = emp_df[emp_df['z_Salary'].abs() > 1]

print("Age Outliers:")
print(age_outliers[['Age', 'z_Age']])

print("\nSalary Outliers:")
print(sal_outliers[['Salary', 'z_Salary']])
```
